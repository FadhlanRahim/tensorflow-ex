# Modeling with tensorflow
1. create a model - define the input and output layer, as well as the hidden layer
2. compile the model - loss function, optimizers, evaluation matrix
3. fit the model - try to find the pattern in our data, X and y, feature and label

````python
tf.random.set_seed(42)

# Create model
model = tf.keras.Sequential([
  tf.keras.layers.Dense(1)
])

# Compile
model.compile(loss = tf.keras.losses.mae,
              optimizer = "SGD",
              metrics = ["mae"])

# fit
model.fit(X, y, epochs=5)
````

### Splitting data
- Training Set : This is what you use to train your model
- Validation Set : This is what you use to tweak your model
- Test Set : This is what you use to evaluate your model

````python
# Split your data
X_train = X[:40]
X_test = X[40:]
y_train = y[:40]
y_test = y[40:]

len(X_train), len(X_test)
````

````python
plt.figure(figsize = (10, 7))

plt.scatter(X_train, y_train, label='train data',c='r')
plt.scatter(X_test, y_test, label='testing data', c='g')
plt.legend
````

````python
tf.random.set_seed(42)

# Create model
model = tf.keras.Sequential([
  tf.keras.layers.Dense(10, input_shape = [1], name='Input'),
  tf.keras.layers.Dense(1, name = 'Output')
])

# Compile
model.compile(loss = "mae",
              optimizer = tf.keras.optimizers.SGD(),
              metrics = ["mae"])
````

````python
model.summary()
````

````python
model.fit(X_train, y_train, epochs=100)
````

- Total Params - number of parameters in the model
- Trainable params - parameters that model can update, trainable
- untrainable-params - freezed parameters, model cannot update these.

````python
# Visualize the model
from tensorflow.keras.utils import plot_model

plot_model(model, show_shapes=True)
````

# Visualizing model's prediction

Plot actual label vs prediction label

````python
y_pred = model.predict(X_test)
y_pred
````

````python
# create plotting function incase we going to use it again
def predictionplot(train_data=X_train, train_labels=y_train, test_data=X_test, test_labels=y_test, predictions=y_pred):
  plt.figure(figsize=(10, 7))
  plt.scatter(train_data, train_labels, c='b', label='training data')
  plt.scatter(test_data, test_labels, c='g', label='testing data')
  plt.scatter(test_data, predictions, c='r', label='predictions')
  plt.legend()
````

````python
predictionplot()
````

## Evaluation Metrics

two common regression metrics :
- MAE - mean absolute error, on average how wrong is the predictions
- MSE - mean square error, squared MAE, blowing up the outliers

````python
model.evaluate(X_test, y_pred)
````

````python
model.evaluate(y_test, y_pred)
````

````python
mae = tf.metrics.mean_absolute_error(y_test, tf.squeeze(y_pred))
mse = tf.metrics.mean_squared_error(y_test, tf.squeeze(y_pred))
mae, mse
````

````python
def mae(y_true, y_pred):
  return tf.metrics.mean_absolute_error(y_true=y_true, y_pred=tf.squeeze(y_pred))

def mse(y_true, y_pred):
  return tf.metrics.mean_squared_error(y_true=y_true, y_pred=tf.squeeze(y_pred))
````

### Normalizing

````python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
ct = make_column_transformer(
    (MinMaxScaler(), ['age', 'bmi', 'children']),
    (OneHotEncoder(handle_unknown='ignore'),['sex', 'smoker', 'region'])
)
````

````python
X1 = insurance.drop('charges', axis=1)
y1 = insurance['charges']
X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size = 0.2, random_state = 42)
````

````python
#fit column transformer to our data
ct.fit(X_train)
#transform train and test
X_train_normalized = ct.transform(X_train)
X_test_normalized = ct.transform(X_test)
````

````python
#build model based on normalized data

tf.random.set_seed(42)
#Build Model
insurancemodel3 = tf.keras.Sequential([
                                  tf.keras.layers.Dense(100),
                                  tf.keras.layers.Dense(10),
                                  tf.keras.layers.Dense(1)
])
#Compile
insurancemodel3.compile(loss="mae",
                        optimizer=tf.keras.optimizers.Adam(0.001),
                        metrics=['mae'])
#fit
history3 = insurancemodel3.fit(X_train_normalized, y_train, epochs=200)
````

````python
insurancemodel3.evaluate(X_test_normalized, y_test)
````

````python
#build model based on normalized data

tf.random.set_seed(42)
#Build Model
insurancemodel4 = tf.keras.Sequential([
                                  tf.keras.layers.Dense(64),
                                  tf.keras.layers.Dense(32),
                                  tf.keras.layers.Dense(16),
                                  tf.keras.layers.Dense(1)
])
#Compile
insurancemodel4.compile(loss="mae",
                        optimizer=tf.keras.optimizers.Adam(0.001),
                        metrics=['mae'])
#fit
history4 = insurancemodel4.fit(X_train_normalized, y_train, epochs=300)
````

````python
insurancemodel4.evaluate(X_test_normalized, y_test)
````
